# Gráficos

## Carregar Bibliotecas

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

## Leitura dos Dados - Gráfico 1

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

```python
# Utiliza valores simples
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000)

# Ordene as Barras
df = df.sort_values(by = 'arrecadacao', ascending = False)

# Utilize Linhas Verticais 
sns.set_style("whitegrid")

# Utilize uma mesma cor para as barras e utilize barras horizontais
plt.figure(figsize = (5,5))
ax = sns.barplot(data = df, x = 'arrecadacao_milhao', y = 'titulo', color = 'tomato')

# Adicione Valores nas Barras
ax.bar_label(
    ax.containers[0],
    fmt = '%.1f',
    label_type = 'center',
    color = 'white'
)

# Melhore o título e os labels dos eixos
plt.xlabel('Arrecadação (Milhões de dólares)')
plt.ylabel('Nome do Filme')
plt.title('Arrecadação de Filmes')

# Adicionar uma linha vertical na origem e remover as bordas
ax.set(frame_on = False)
ax.axvline(0, color = 'k', clip_on = False)
plt.show()
```

## Leitura dos Dados - Gráfico 2

```python
df2 = sns.load_dataset("penguins")
df2.head()
```

```python
# Utilieza linhas verticais e horizontais
sns.set_style("whitegrid")

# Melhore o título e os labels dos eixos
ax = sns.histplot(data = df2, x = 'body_mass_g')
plt.ylabel('Quantidade')
plt.xlabel('Massa(g)') 
plt.title('Distribuição da massa dos pinguins')

# Adicionar uma linha a origem e remover as bordas
ax.set(frame_on = False)
ax.axvline(0, color = 'k', clip_on = False)
plt.show()
```

## Leitura dos Dados - Gráfico 3

```python
df3 = sns.load_dataset("tips")
df3.head()
```

```python
# Utilize linhas verticais
sns.set_style("whitegrid")

# Utilize uma mesma cor para as barras
plt.figure(figsize = (6, 3))
ax = sns.boxplot(x = "day", y = "total_bill", data = df3, color = 'cornflowerblue')

# Melhore o título e os labels dos eixos
plt.ylabel('Dia da semana')
plt.xlabel('Conta (USD)') 
plt.title('Distribuição de gorjetas ao longo da semana')
plt.show()
```
